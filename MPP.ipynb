```python
import re 
def backCamera_MP(x):
    matching_pattern = re.search(r'(\d+) MP' ,x)
    return matching_pattern.group(1) if matching_pattern else None
def frontCamera_MP(x):
    matching_pattern = re.search(r'&amp;\s*(\d+)\s*MP', x)
    return matching_pattern.group(1) if matching_pattern else None


def Numberof_RearCamera(x):
    matching_pattern = re.search(r'(Dual|Triple|Quad) Rear',x)
    return matching_pattern.group(1) if matching_pattern else None

df['Back Camera MP'] = df['Camera'].apply(backCamera_MP)
df['Front Camera MP'] = df['Camera'].apply(frontCamera_MP)
df['Number of Rear Camera'] = df['Camera'].apply(Numberof_RearCamera)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pandas as pd



categorical_columns = ['Brand', 'Processor']
encoding = OneHotEncoder(sparse_output=False).set_output(transform='pandas')
transformed = encoding.fit_transform(X[categorical_columns])

# one hot encoding only to categorical columns and concating with the original dataframe
concated_newdf = pd.concat([X, transformed], axis=1).drop(columns=categorical_columns)

# target variable in Y
Y = concated_newdf['Price']
X = concated_newdf.drop(columns='Price')

# scaled numerical columns
scaler1 = StandardScaler()
X[['Spec_score', 'Ram', 'Battery', 'Display', 'Inbuilt_memory', 'Width', 'Height', 
   'Back Camera MP', 'Front Camera MP', 'Number of Rear Camera', 'PPI']] = \
    scaler1.fit_transform(X[['Spec_score', 'Ram', 'Battery', 'Display', 'Inbuilt_memory',
                            'Width', 'Height', 'Back Camera MP', 'Front Camera MP', 
                            'Number of Rear Camera', 'PPI']])


print(X.shape, Y.shape)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import learning_curve


def plot_learning_curve(model, X, Y, title):
    
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    train_sizes, train_scores, test_scores = learning_curve(
        model, X, Y, cv=kf, scoring="r2", train_sizes=np.linspace(0.1, 1.0, 10))
    
  
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    plt.figure(figsize=(8, 5))
    plt.plot(train_sizes, train_mean, "o-", color="blue", label="Training Score")
    plt.plot(train_sizes, test_mean, "o-", color="red", label="Validation Score")
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color="blue")
    plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.1, color="red")
    plt.ylim(-1.2 , 1.2)
    plt.title(f"Learning Curve: {title}")
    plt.xlabel("Training Set Size")
    plt.ylabel("R Score")
    plt.legend(loc="best")
    plt.show()
```

```python
def train_model(model, X, Y):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,  shuffle= True, random_state=42)
       
    model.fit(X_train, Y_train)
    
    Y_pred = model.predict(X_test)
    
    mse = mean_squared_error(Y_test, Y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(Y_test, Y_pred)
    
    print(f"Root Mean Squared Error (RMSE): {rmse}")
    print(f"R Score: {r2}")
    
    residuals = Y_test - Y_pred
    plt.scatter(Y_test, residuals)
    plt.axhline(y=0, color='r', linestyle='--')
    plt.xlabel("True Values")
    plt.ylabel("Residuals")
    plt.title("Residual Plot")
    plt.show()
    
    cv_scores = cross_val_score(model, X, Y, cv=5, scoring='neg_mean_squared_error')
    cv_rmse = np.sqrt(-cv_scores)
    
    
    cv_r2_scores = cross_val_score(model, X, Y, cv=5, scoring='r2')
    
    print(f"Cross-validated RMSE: {cv_rmse.mean()}")
    print(f"Cross-validated R Score: {cv_r2_scores.mean()}")
    
    plot_learning_curve(model,X_train,Y_train,(model))
    return r2,rmse
    
    
```

```python
linear_r2, linear_rmse = train_model(LinearRegression(), X, Y)
```

cross validation MSE and training MSE have been almost same after 500 600 data so, model heavily underfits the data. As adding more data does not improve the generalization.

```python
import matplotlib.pyplot as plt

def plot_results(results):
   
    plt.figure(figsize=(12, 6))
    
    # Plot R^2 vs alpha
    plt.subplot(1, 3, 1)
    plt.plot(results['alpha'], results['r2'], 'o-', color = 'blue',label='R Score')
   
    plt.xlabel('Alpha')
    plt.ylabel('R Score')
    plt.title('R vs Alpha')
    
    
    # Plot mse vs alpha
    plt.subplot(1, 3, 2)
    plt.plot(results['alpha'], results['mse'],'o-', color = 'red', label='MSE')
    
    plt.xlabel('Alpha')
    plt.ylabel('Mean Squared Error')
    plt.title('MSE vs Alpha')
    
    # Plot mae vs alpha
    plt.subplot(1, 3, 3)
    plt.plot(results['alpha'], results['mae'], 'o-', color = 'green',label='MAE')
    
    plt.xlabel('Alpha')
    plt.ylabel('Mean Absolute Error')
    plt.title('MAE vs Alpha')
    
    plt.tight_layout()
    plt.show()
```

```python


from sklearn.linear_model import Ridge, Lasso

def train_model_with_alpha_cv(model_class, X, Y, alphas=np.linspace(1, 100, 10)):
    best_alpha = None
    best_r2 = -np.inf
    best_mse = np.inf
    best_mae = np.inf
    results = {'alpha': [], 'r2': [], 'mse': [], 'mae': []}
  
    for value in alphas:
        if model_class == 'ridge':
         model = Ridge(value)
        elif model_class == 'lasso':
         model = Lasso(alpha=50*value, max_iter=10000)
        else: 
            print("Value Error")
      
        r2_scores = cross_val_score(model, X, Y, cv=5, scoring='r2')
        mse_scores = cross_val_score(model, X, Y, cv=5, scoring='neg_mean_squared_error')
        mae_scores = cross_val_score(model, X, Y, cv=5, scoring='neg_mean_absolute_error')
        
        mean_r2 = r2_scores.mean()
        mean_mse = -mse_scores.mean() 
        mean_mae = -mae_scores.mean()  
       
        if(model_class == 'lasso'):
          results['alpha'].append(value* 50 )
          
        else: 
          results['alpha'].append(value)
          
        results['r2'].append(mean_r2)
        results['mse'].append(mean_mse)
        results['mae'].append(mean_mae)
        
        # best model
        if mean_r2 > best_r2:
            best_r2 = mean_r2
            if(model_class=='lasso'):
              best_alpha = value *50
            else: 
              best_alpha = value
        if mean_mse < best_mse:
            best_mse = mean_mse
        if mean_mae < best_mae:
            best_mae = mean_mae

   
    plot_results(results)
   
   
    print(f"Best Alpha: {best_alpha}")
    print(f"Best R Score: {best_r2}")
    print(f"Best MSE: {best_mse}")
    print(f"Best MAE: {best_mae}")
    
    return best_r2, best_mse ** 0.5
```

```python
from sklearn.preprocessing import PolynomialFeatures

numerical_columns = X.columns[:11] 

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(concated_newdf[numerical_columns])

scaler = StandardScaler()
X_poly_scaled = scaler.fit_transform(X_poly)

X_remaining = X.iloc[:, 11:]
X_poly_final = np.concatenate((X_poly_scaled, X_remaining), axis=1)
```

```python
ridge_r2, ridge_rmse = train_model_with_alpha_cv('ridge', X_poly_final, Y)
```

```python
lasso_r2, lasso_rmse = train_model_with_alpha_cv('lasso', X_poly_final  , Y)
```

```python

from sklearn.model_selection import  GridSearchCV
from sklearn.metrics import mean_absolute_error

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
rf = RandomForestRegressor(random_state=42)
param_grid = {
    'n_estimators': [200, 300, 500],  # Number of trees
    'max_depth': [10, 20, None],  # Depth of each tree
    'min_samples_split': [2, 5, 10],  # Minimum samples to split a node
    'min_samples_leaf': [1, 2, 4],  # Minimum samples in each leaf
    'max_features': ['sqrt', 'log2']  # Number of features to consider at each split
}

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, 
                           cv=5, scoring='r2', n_jobs=-1, verbose=2)

grid_search.fit(X_train, y_train)

best_rf = grid_search.best_estimator_
print(f"Best Parameters: {grid_search.best_params_}")

y_pred_best = best_rf.predict(X_test)

mae = mean_absolute_error(y_test, y_pred_best)
mse = mean_squared_error(y_test, y_pred_best)
best_model_rmse = np.sqrt(mse)
best_model_r2 = r2_score(y_test, y_pred_best)

print(f"\nModel Performance:")
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {best_model_rmse:.4f}")
print(f"R Score: {best_model_r2:.4f}")
```
